## Section 2. Description of the Dataset

### Data Collection Process

- **Coverage:**  
  The survey has nationwide coverage, including all 17 administrative regions across the Philippines' archipelagic geography.

- **Target Population:**  
  All households in the Philippines, with labor force information collected from household members aged 15 years and above.

- **Sampling Design:**  
  - **Stage 1:** Selection of Primary Sampling Units (PSUs) using systematic sampling with probability proportional to size (PPS).  
  - **Stage 2:** Selection of housing units/households within sampled PSUs using systematic sampling.

- **Sample Size:**  
  Typically around 42,000 to 43,000 households nationwide.

- **Data Collection Method:**  
  Trained PSA field staff conduct face-to-face interviews with household members using structured questionnaires. The interviews follow standardized protocols to ensure data quality and comparability.

- **Quality Control:**  
  Field supervisors conduct spot checks and re-interviews on a subset of surveyed households to ensure data integrity and consistency.

````python
# @title Neural Network.py


import torch.nn as nn
import torch.nn.init


class NeuralNetwork(nn.Module):

    def __init__(self,
                 input_size,
                 num_classes,
                 list_hidden,
                 activation='relu'):
        """Class constructor for NeuralNetwork

        Arguments:
            input_size {int} -- Number of features in the dataset
            num_classes {int} -- Number of classes in the dataset
            list_hidden {list} -- List of integers representing the number of
            units per hidden layer in the network
            activation {str, optional} -- Type of activation function. Choices
            include 'sigmoid', 'tanh', and 'relu'.
        """
        super(NeuralNetwork, self).__init__()

        self.input_size = input_size
        self.num_classes = num_classes
        self.list_hidden = list_hidden
        self.activation = activation


    def create_network(self):
        layers = []

        # Input + Hidden Layers
        layers.append(nn.Linear(self.input_size, self.list_hidden[0]))
        layers.append(self.get_activation(self.activation))  # ReLU/tanh for hidden

        for i in range(len(self.list_hidden) - 1):
            layers.append(nn.Linear(self.list_hidden[i], self.list_hidden[i+1]))
            layers.append(self.get_activation(self.activation))

        # Output: 1 unit with sigmoid
        layers.append(nn.Linear(self.list_hidden[-1], 1))
        # layers.append(nn.Sigmoid()) # remove cus we using bcewithlogitsloss

        self.layers = nn.Sequential(*layers)

    def init_weights(self):
        """Initializes the weights of the network. Weights of a
        torch.nn.Linear layer should be initialized from a normal
        distribution with mean 0 and standard deviation 0.1. Bias terms of a
        torch.nn.Linear layer should be initialized with a constant value of 0.
        """
        torch.manual_seed(2)

        # For each layer in the network
        for module in self.modules():

            # If it is a torch.nn.Linear layer
            if isinstance(module, nn.Linear):

                # TODO: Initialize the weights of the torch.nn.Linear layer
                # from a normal distribution with mean 0 and standard deviation
                # of 0.1.
                # HINT: Use nn.init.normal_() function.
                nn.init.kaiming_normal_(module.weight, mode="fan_in", nonlinearity="relu")

                # TODO: Initialize the bias terms of the torch.nn.Linear layer
                # with a constant value of 0.
                # HINT: Use nn.init.constant_() function.
                nn.init.constant_(module.bias, 0)

    def get_activation(self,
                       mode='sigmoid'):
        """Returns the torch.nn layer for the activation function.

        Arguments:
            mode {str, optional} -- Type of activation function. Choices
            include 'sigmoid', 'tanh', and 'relu'.

        Returns:
            torch.nn -- torch.nn layer representing the activation function.
        """
        activation = nn.Sigmoid()

        if mode == 'tanh':
            activation = nn.Tanh()

        elif mode == 'relu':
            activation = nn.ReLU()

        return activation

    def forward_manual(self,
                       x,
                       verbose=False):
        """Forward propagation of the model, implemented manually.

        Arguments:
            x {torch.Tensor} -- A Tensor of shape (N, D) representing input
            features to the model.
            verbose {bool, optional} -- Indicates if the function prints the
            output or not.

        Returns:
            torch.Tensor, torch.Tensor -- A Tensor of shape (N, C) representing
            the output of the final linear layer in the network. A Tensor of
            shape (N, C) representing the probabilities of each class given by
            the softmax function.
        """

        # For each layer in the network
        for i in range(len(self.layers) - 1):

            # If it is a torch.nn.Linear layer
            if isinstance(self.layers[i], nn.Linear):

                # TODO: Compute the result of the linear layer. Do not forget
                # to add the bias term. Assign the result to x.
                # HINT: Use torch.matmul() function.
                x = torch.matmul(x, self.layers[i].weight.T) + self.layers[i].bias

            # If it is another function
            else:
                # Call the forward() function of the layer
                # and return the result to x.
                x = self.layers[i](x)

            if verbose:
                # Print the output of the layer
                print('Output of layer ' + str(i))
                print(x, '\n')

        # Apply the softmax function
        probabilities = self.layers[-1](x)

        if verbose:
            print('Output of layer ' + str(len(self.layers) - 1))
            print(probabilities, '\n')

        return x, probabilities

    def forward(self,
                x,
                verbose=False):
        """Forward propagation of the model, implemented using PyTorch.

        Arguments:
            x {torch.Tensor} -- A Tensor of shape (N, D) representing input
            features to the model.
            verbose {bool, optional} -- Indicates if the function prints the
            output or not.

        Returns:
            torch.Tensor, torch.Tensor -- A Tensor of shape (N, C) representing
            the output of the final linear layer in the network. A Tensor of
            shape (N, C) representing the probabilities of each class given by
            the softmax function.
        """


        x = self.layers(x)  # Ensure the entire sequential model runs

        if verbose:
            print(f'Final logits output:\n{x}\n')

        return x

    def predict(self,
                probabilities, threshold=0.5):
        """Returns the index of the class with the highest probability.

        Arguments:
            probabilities {torch.Tensor} -- A Tensor of shape (N, C)
            representing the probabilities of N instances for C classes.

        Returns:
            torch.Tensor -- A Tensor of shape (N, ) contaning the indices of
            the class with the highest probability for N instances.
        """

        # TODO: Return the index of the class with the highest probability
        return (probabilities > threshold).float().squeeze()  # Converts to 0/1
````

````python
# @title DataLoader.py
import numpy as np


class DataLoader(object):

    def __init__(self, X, y, batch_size):
        """Class constructor for DataLoader

        Arguments:
            X {np.ndarray} -- A numpy array of shape (N, D) containing the
            data; there are N samples each of dimension D.
            y {np.ndarray} -- A numpy array of shape (N, 1) containing the
            ground truth values.
            batch_size {int} -- An integer representing the number of instances
            per batch.
        """
        self.X = X
        self.y = y
        self.batch_size = batch_size

        self.indices = np.array([i for i in range(self.X.shape[0])])
        np.random.seed(1)

    def shuffle(self):
        """Shuffles the indices in self.indices.
        """

        # TODO: Use np.random.shuffle() to shuffles the indices in self.indices
        np.random.shuffle(self.indices)

    def get_batch(self, mode='train'):
        """Returns self.X and self.y divided into different batches of size
        self.batch_size according to the shuffled self.indices.

        Arguments:
            mode {str} -- A string which determines the mode of the model. This
            can either be `train` or `test`.

        Returns:
            list, list -- List of np.ndarray containing the data divided into
            different batches of size self.batch_size; List of np.ndarray
            containing the ground truth labels divided into different batches
            of size self.batch_size
        """

        X_batch = []
        y_batch = []

        # TODO: If mode is set to `train`, shuffle the indices first using
        # self.shuffle().
        if mode == 'train':
            self.shuffle()
        elif mode == 'test':
            self.indices = np.array([i for i in range(self.X.shape[0])])

        # The loop that will iterate from 0 to the number of instances with
        # step equal to self.batch_size
        for i in range(0, len(self.indices), self.batch_size):

            # TODO: Check if we can still get self.batch_size from the
            # remaining indices starting from index i. Edit the condition
            # below.
            if i + self.batch_size <= len(self.indices):
                indices = self.indices[i:i + self.batch_size]

            # TODO: Else, just get the remaining indices from index i until the
            # last element in the list. Edit the statement inside the else
            # block.
            else:
                indices = self.indices[i:]

            X_batch.append(self.X[indices])
            y_batch.append(self.y[indices])

        return X_batch, y_batch
````

Considering the missing values, we will do **prediction of employment status**.

This classification task will determine whether a person is employed, unemployed, not in labor force based on demographics, education and location.

For this, the target variable or label would be the Employment status.

- Employed	-> 1
- Not Employed -> 2 (Combined from ```Unemployed``` and ```Not in the labor force```)

Since, the dataset is largely imbalanced. With attempts to do balancing using synthetic data didn't show good results.
We opted to do binary classification instead, merging ```Unemployed``` and ```Not in the labor force```.

For example, using "hours worked in the past week" directly implies employment, making it an unsuitable predictor.

Thus, this classification task focuses on demographic and non-job-related factors to predict whether a person is employed or not.

````python
# @title old_onehotencode
def one_hot_encode_features(X_train, X_val, X_test, categorical_features,num_features):
  # Apply One-Hot Encoding to categorical features
  encoder = OneHotEncoder(drop="first", sparse_output=False, handle_unknown="ignore")
  X_train_encoded = encoder.fit_transform(X_train[categorical_features])
  X_val_encoded = encoder.transform(X_val[categorical_features])
  X_test_encoded = encoder.transform(X_test[categorical_features])

  # Convert back to DataFrame with proper names
  feature_names = encoder.get_feature_names_out(categorical_features)
  X_train_encoded = pd.DataFrame(X_train_encoded, columns=feature_names, index=X_train.index)
  X_val_encoded = pd.DataFrame(X_val_encoded, columns=feature_names, index=X_val.index)
  X_test_encoded = pd.DataFrame(X_test_encoded, columns=feature_names, index=X_test.index)

  # Drop original categorical columns and concatenate new one-hot encoded columns
  X_train = pd.concat([X_train.drop(columns=categorical_features), X_train_encoded], axis=1)
  X_val = pd.concat([X_val.drop(columns=categorical_features), X_val_encoded], axis=1)
  X_test = pd.concat([X_test.drop(columns=categorical_features), X_test_encoded], axis=1)

  # Apply MinMaxScaler to numerical features
  scaler = MinMaxScaler()
  X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train[num_features]),
                                columns=num_features, index=X_train.index)
  X_val_scaled = pd.DataFrame(scaler.fit_transform(X_val[num_features]),
                                columns=num_features, index=X_val.index)
  X_test_scaled = pd.DataFrame(scaler.transform(X_test[num_features]),
                              columns=num_features, index=X_test.index)

  # Replace original numerical features with scaled values
  X_train[num_features] = X_train_scaled
  X_val[num_features] = X_val_scaled
  X_test[num_features] = X_test_scaled

  X_train = X_train.apply(pd.to_numeric)
  X_val = X_val.apply(pd.to_numeric)
  X_test = X_test.apply(pd.to_numeric)

  return X_train, X_val, X_test
````

````python
# @title confusion_matrix_network
def confusion_matrix_network(network, X_test, X_train, y_test, y_train):
  logits = network.forward(X_test)
  probabilities = torch.sigmoid(logits)
  predictions = network.predict(probabilities)
  y_pred = predictions
  cm = confusion_matrix(y_test, y_pred)
  sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Employed","Not Employed"], yticklabels=["Employed","Not Employed"],)
  plt.xlabel("Predicted")
  plt.ylabel("Actual")
  plt.show()
````

````python
# @title def network_eval()
def network_eval(network):
  network.eval()
  logits = network.forward(X_test)
  probabilities = torch.sigmoid(logits)
  predictions = network.predict(probabilities)
  print(predictions)
  num_correct = (predictions == y_test).sum().item()
  print(num_correct)
  accuracy = num_correct / len(predictions)
  print('{:.4f}'.format(accuracy))
````

````python
# @title classification_report_network
def classification_report_network(network, X_test, X_train, y_test, y_train):
  y_true = y_test
  logits = network.forward(X_test)
  probabilities = torch.sigmoid(logits)
  predictions = network.predict(probabilities)
  y_pred = predictions
  target_names = ['0', '1']
  print("Test")
  print(classification_report(y_true, y_pred, target_names=target_names))
  y_true = y_train
  logits = network.forward(X_train)
  probabilities = torch.sigmoid(logits)
  predictions = network.predict(probabilities)
  y_pred = predictions
  print("TRAIN")
  print(classification_report(y_true, y_pred, target_names=target_names))
````

````python
# @title train_nn_model
def train_nn_model(model, criterion, optimizer, data_loader, X_val, y_val, max_epochs=400, patience=50):
    best_val_loss = float('inf')
    patience_counter = 0
    losses = []
    val_losses = []

    for epoch in range(max_epochs):
        # Training phase
        current_epoch_loss = 0
        X_batch, y_batch = data_loader.get_batch()

        for X, y in zip(X_batch, y_batch):
            X = torch.as_tensor(X, dtype=torch.float32)
            y = torch.as_tensor(y, dtype=torch.float32)

            optimizer.zero_grad()
            logits = model(X)
            loss = criterion(logits.squeeze(), y)

            loss.backward()
            optimizer.step()

            current_epoch_loss += loss.item()

        average_loss = current_epoch_loss / len(X_batch)
        losses.append(average_loss)

        # Validation phase
        with torch.no_grad():
            model.eval()  # Switch to evaluation mode
            val_logits = model(torch.as_tensor(X_val, dtype=torch.float32))
            val_loss = criterion(val_logits.squeeze(), torch.as_tensor(y_val, dtype=torch.float32))
            model.train()  # Switch back to training mode

        val_losses.append(val_loss.item())
        print(f'Epoch: {epoch + 1} | Train Loss: {average_loss:.6f} | Val Loss: {val_loss.item():.6f}')

        # Early stopping (optional)
        if val_loss.item() < best_val_loss - 1e-6:
            best_val_loss = val_loss.item()
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch + 1}")
            break

    return losses, val_losses
````

````python
def describe_tree(clf):
    n_nodes = clf.tree_.node_count
    children_left = clf.tree_.children_left
    children_right = clf.tree_.children_right
    feature = clf.tree_.feature
    threshold = clf.tree_.threshold
    values = clf.tree_.value

    node_depth = np.zeros(shape=n_nodes, dtype=np.int64)
    is_leaves = np.zeros(shape=n_nodes, dtype=bool)
    stack = [(0, 0)]  # start with the root node id (0) and its depth (0)
    while len(stack) > 0:
        # `pop` ensures each node is only visited once
        node_id, depth = stack.pop()
        node_depth[node_id] = depth

        # If the left and right child of a node is not the same we have a split
        # node
        is_split_node = children_left[node_id] != children_right[node_id]
        # If a split node, append left and right children and depth to `stack`
        # so we can loop through them
        if is_split_node:
            stack.append((children_left[node_id], depth + 1))
            stack.append((children_right[node_id], depth + 1))
        else:
            is_leaves[node_id] = True

    print(
        "The binary tree structure has {n} nodes and has "
        "the following tree structure:\n".format(n=n_nodes)
    )
    for i in range(n_nodes):
        if is_leaves[i]:
            print(
                "{space}node={node} is a leaf node, values: {values}.".format(
                    space=node_depth[i] * "\t", node=i, values=values[i]
                )
            )
        else:
            print(
                "{space}node={node} is a split node: "
                "go to node {left} if X[:, {feature}] <= {threshold} "
                "else to node {right}.".format(
                    space=node_depth[i] * "\t",
                    node=i,
                    left=children_left[i],
                    feature=feature[i],
                    threshold=threshold[i],
                    right=children_right[i],
                )
            )
````

Function that categorizes a grade into education levels: "Primary," "Secondary," "College," or "Postgraduate." If the grade is invalid, it returns "Unknown."

````python
# @title Categorize_education
# Function to categorize education levels
def categorize_education(grade):
    try:
        grade = int(grade)  # Convert to integer safely
    except ValueError:
        return "Unknown"  # Handle unexpected values

    if grade < 280:
        return "Primary"
    elif 280 <= grade < 350:
        return "Secondary"
    elif 350 <= grade < 810:
        return "College"
    else:
        return "Postgraduate"
````

````python
# @title train_nn_model2
def train_nn_model(model, criterion, optimizer, data_loader, X_val, y_val, max_epochs=400, patience=50, trial=None):
    best_val_loss = float('inf')
    patience_counter = 0
    losses = []
    val_losses = []

    for epoch in range(max_epochs):
        # Training phase
        current_epoch_loss = 0
        X_batch, y_batch = data_loader.get_batch()

        for X, y in zip(X_batch, y_batch):
            X = torch.as_tensor(X, dtype=torch.float32)
            y = torch.as_tensor(y, dtype=torch.float32)

            optimizer.zero_grad()
            logits = model(X)
            loss = criterion(logits.squeeze(), y)

            loss.backward()
            optimizer.step()

            current_epoch_loss += loss.item()

        average_loss = current_epoch_loss / len(X_batch)
        losses.append(average_loss)

        # Validation phase
        with torch.no_grad():
            model.eval()  # Switch to evaluation mode
            val_logits = model(torch.as_tensor(X_val, dtype=torch.float32))
            val_loss = criterion(val_logits.squeeze(), torch.as_tensor(y_val, dtype=torch.float32))
            model.train()  # Switch back to training mode

        val_losses.append(val_loss.item())

        # Log progress
        print(f"Epoch {epoch + 1}: Train Loss = {average_loss:.6f}, Val Loss = {val_loss.item():.6f}")

        # Report intermediate result for pruning
        if trial is not None:
            trial.report(val_loss.item(), epoch)

            # Check if the trial should be pruned
            if trial.should_prune():
                print(f"Trial pruned at epoch {epoch + 1} due to high validation loss.")
                raise optuna.TrialPruned()

        # Early stopping (optional)
        if val_loss.item() < best_val_loss - 1e-6:
            best_val_loss = val_loss.item()
            patience_counter = 0
        else:
            patience_counter += 1

        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch + 1}")
            break

    return losses, val_losses
````

The objective function defines the workflow for training and validating the neural network. It leverages Optuna's hyperparameter suggestion methods to explore and evaluate configurations for the following parameters:

Learning Rate: The learning rate was sampled logarithmically using trial.suggest_loguniform. This method allows testing both smaller, stable values for gradual learning and larger values for faster convergence.

Hidden Layers: Three hidden layer configurations were tested:

A single hidden layer with the same size as the input features.

Two hidden layers with progressively smaller sizes.

Three hidden layers, with sizes decreasing further in each successive layer.

Activation Function: The ReLU activation function was selected as the primary focus for optimization using trial.suggest_categorical. This choice aligns with best practices in deep learning due to ReLU's effectiveness in mitigating the vanishing gradient problem and promoting faster training convergence.

Patience: The early stopping patience, which determines the number of epochs without improvement before stopping training, was varied between 10 and 50 epochs to balance training time and performance.


````python
import optuna
from optuna.pruners import MedianPruner

def objective(trial):
    print(f"\nStarting Trial {trial.number}")

    batch_size = 32
    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1e-1, log=True)
    hidden_layers = trial.suggest_categorical(
        'hidden_layers',
        [
            (X_train.shape[1],),
            (X_train.shape[1], int(X_train.shape[1] / 2)),
            (X_train.shape[1], int(X_train.shape[1] / 2), int(X_train.shape[1] / 4))
        ]
    )
    activation = trial.suggest_categorical('activation', ['relu'])
    patience = trial.suggest_int('patience', 10, 50)

    print(f"Trial {trial.number}: Learning Rate = {learning_rate}, Hidden Layers = {hidden_layers}, "
          f"Activation = {activation}, Patience = {patience}")

    data_loader = DataLoader(X_train, y_train, batch_size=batch_size)

    network = NeuralNetwork(
        input_size=X_train.shape[1],
        num_classes=1,
        list_hidden=hidden_layers,
        activation=activation,
    )
    network.create_network()
    network.init_weights()

    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.Adam(network.parameters(), lr=learning_rate)

    # Train Model with Pruning
    try:
        losses, val_losses = train_nn_model(
            network, criterion, optimizer, data_loader,
            X_val, y_val, max_epochs=400, patience=patience, trial=trial
        )
    except optuna.TrialPruned:
        print(f"Trial {trial.number} was pruned.")
        return float('inf')  # Return a large value to indicate failure

    # Return the best validation loss
    best_val_loss = min(val_losses)
    print(f"Trial {trial.number} completed with best validation loss: {best_val_loss:.6f}")
    return best_val_loss
````

````python
study = optuna.create_study(
    direction='minimize',
    pruner=MedianPruner(n_startup_trials=5, n_warmup_steps=30, interval_steps=10)
)

print("Starting optimization...")
study.optimize(objective, n_trials=20)  # Perform 20 trials

# Print the best parameters and validation loss
print("\nOptimization complete!")
print("Best Parameters:", study.best_params)
print("Best Validation Loss:", study.best_value)
````

### Model 3: Random Forest Optimization

For the random forest model, we'll be using GridSearchCV to find the best combination of parameters.

#### Hyperparameter Tuning using GridSearch

````python
# Define hyperparameters to test
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5, n_jobs=-1, verbose=2)

grid_search.fit(X_train, y_train)
print("Best Parameters:", grid_search.best_params_)
````

Apply the best parameters to the learning model.

````python
rf_best = RandomForestClassifier(
    max_depth=10,
    min_samples_leaf=1,
    min_samples_split=5,
    n_estimators=100,
    random_state=42
)

rf_best.fit(X_train, y_train)
````

````python
predictions = rf_best.predict(X_test)
num_correct = (predictions == y_test).sum().item()
accuracy = num_correct / len(predictions)
print(f"Accuracy: {accuracy:.4f}\n")

print(classification_report(y_test, predictions))
````

#### Learning from Model Performance:

- **Neural Networks:** These models excel at capturing complex patterns, but their computational demands and sensitivity to hyperparameter tuning highlight the need for strategic resource allocation and experimentation during training.

#### Impact of Optimization:
- **Hyperparameter tuning** demonstrated how small adjustments can lead to significant improvements in performance, especially in metrics like recall and F1-score. For example, increasing the maximum depth or adjusting the number of estimators improved Random Forest's ability to generalize across the dataset.

#### Handling Class Imbalance:
- The challenges observed with minority classes (e.g., Class 2) underscore the importance of addressing imbalanced datasets. Precision improvements in optimized models illustrate how balancing techniques and careful tuning can reduce bias toward majority classes.

#### Understanding Metrics Beyond Accuracy:
- Metrics like **precision, recall, and F1-score** were critical in evaluating performance beyond overall accuracy. This learning reinforces the importance of aligning evaluation metrics with real-world requirements, such as prioritizing recall in applications where missing a positive instance has a higher cost.
